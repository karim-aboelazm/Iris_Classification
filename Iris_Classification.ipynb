# Dataset

We will use [IRIS dataset](http://archive.ics.uci.edu/ml/machine-learning-databases/iris/). It has 150 instances with 4 features:
1. Sepal Length
2. Sepal Width
3. Petal Length
4. Petal Width

from 3 classes:
* Iris-setosa
* Iris-versicolor
* Iris-virginica

```python
iris = pd.read_csv('iris.data', names = ['sepal_length', 'sepal_width',\
        'petal_length', 'petal_width', 'label'], skipfooter = 1, engine = 'python')
```

# Machine Learning

Let us divide the dataset into training (70%) and test set(30%). We will use trainset to build our model, and test set to evaluate the perfromance.

```python
from random import sample
idx = sample(iris.index, int(iris.shape[0] * 0.7))
iris_train, iris_test = iris.ix[idx], iris.drop(idx)
```

### Notations:

We will denote the $j^{th}$ predictor variables or features of $i^{th}$ instance with $X_j^{(i)} \in \mathbb R$, number of data instances as $m$, instance of all predictor variables with $X^{(i)} \in \mathbb R^n \ for\ i = 1..m$, where $n$ is number of dimensions.
and $X \in \mathbb R^{m * n}$ denotes whole training set of all instances and features. The target variable is denoted by $y$.

## Preprocessing (Normalize)

Normalize the dataset so that it has zero mean and unit variance.
$$ X_j^{(i)} = \frac{ X_j^{(i)} - \mu_j }{\sigma_j} $$
This is available in **sklearn.preprocessing**.

```python
X_train = np.array(iris_train.as_matrix()[:, :4], dtype = np.float16)
X_test = np.array(iris_test.as_matrix()[:, :4], dtype = np.float16)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
np.mean(X_train), np.std(X_train)
```

## Dimensionality Reduction (PCA)

We will perform PCA to reduce dimensions to 2, so that it will assist visualization. PCA is available in **sklearn.decomposition**.

```python
from sklearn.decomposition import PCA
pca = PCA()
X_train = pca.fit_transform(X_train)
ax = plt.figure(figsize = (5, 3))
plt.plot([1, 2, 3, 4], np.cumsum(pca.explained_variance_ratio_), '-rd')
plt.xlabel("No. of Principal Components")
plt.ylabel("Cumulative variance")
```

From the plot, it is clear that just two principal components contains more than $95\%$ information. So, let us just drop the other components safely and see how it looks.

```python
plt.figure(figsize = (5, 4))
plt.scatter(X_train[:, 0], X_train[:, 1], c = [col_map[lb] for lb in iris.label.ix[idx]], s = 75, alpha = 0.4)
```

## Model #1: Logistic Regression

### Learning

Logistic regression and many others models are available in **scikit-learn**. The below example is brought from [here](http://scikit-learn.org/stable/auto_examples/linear_model/plot_iris_logistic.html).

```python
from sklearn import linear_model
h = .02  # step size in the mesh

logreg = linear_model.LogisticRegression(C=1e5)

# we create an instance of Neighbours Classifier and fit the data.
logreg.fit(X_train[:, :2], np.array(iris.ix[idx].label, dtype = np.str))

# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, m_max]x[y_min, y_max].
x_min, x_max  = X_train[:, 0].min() - .5, X_train[:, 0].max() + .5
y_min, y_max = X_train[:, 1].min() - .5, X_train[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = logreg.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
encoder = {'Iris-setosa': 1, 'Iris-versicolor': 2, 'Iris-virginica': 3}
Z = np.array([encoder[z] for z in Z]).reshape(xx.shape)
plt.figure(figsize=(4, 3))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# Plot also the training points
plt.scatter(X_train[:, 0], X_train[:, 1], c= [col_map[lb] for lb in iris.label.ix[idx]], edgecolors='k', cmap=plt.cm.Paired)
plt.xlabel('comp 1')
plt.ylabel('comp 2')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())

plt.show()
```

### Evaluation

We have fitted our model with training set, now let us check how well it does for test-set. First, let us scale and perform PCA with our fitted model for them with training set. We **should not** scale or perform PCA with test set, but rather use their models from train set to scale and perform PCA.

```python
X_test = pca.transform(scaler.transform(X_test))
y_pred = logreg.predict(X_test[:, :2])
```

```python
from sklearn import metrics
print metrics.classification_report(\
        np.array(iris.label.drop(idx).as_matrix(), dtype = np.str), y_pred)
```
